```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline 
```

```python
df_ted=pd.read_csv('ted_main.csv')
df_ted.head()
```

```python
df_ted['film_date'] = pd.to_datetime(df_ted['film_date'],unit='s')
df_ted['published_date'] = pd.to_datetime(df_ted['published_date'],unit='s')
df_ted.head()
```

```python
df_ted=df_ted[['published_date','film_date','comments','description','duration','event','languages','main_speaker','title','tags','views']]
```

```python
df_ted['year'] = pd.DatetimeIndex(df_ted['film_date']).year
```

```python
df_ted.tags = df_ted.tags.astype(str)
```

```python
df_ted[df_ted.tags.str.contains('science')].groupby('year').count()['title'].plot('bar')
```

```python
df_ted.year.value_counts()
```

```python
tags = pd.DataFrame(df_ted.tags.sum().split(), columns = ['tag'])
tags.tag = tags.tag.str.replace('[','')
tags.tag = tags.tag.str.replace(',','')
tags.tag = tags.tag.str.replace(']','')
tags.tag = tags.tag.str.replace("'",'')
tags_df=pd.DataFrame(tags.tag.value_counts())
tags_df=tags_df.rename(columns={'tag':'aantal'})
tags_df[:20]
```

```python
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS

mpl.rcParams['font.size']=12                 
mpl.rcParams['savefig.dpi']=100              
mpl.rcParams['figure.subplot.bottom']=.1 

data = tags

wordcloud = WordCloud(
                          background_color='white',
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(data['tag']))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```

```python
df_ted['first name'] = df_ted.main_speaker.str.split(' ', expand=True)[0]
```

```python
!pip install gender_guesser
```

```python
import gender_guesser.detector as gender
d = gender.Detector()
data = df_ted['first name'].astype(str)
gender = []
for x in data:
    gender.append(d.get_gender(x))
        
```

```python
df_ted['gender'] = gender
df_ted.gender = df_ted.gender.str.replace('mostly_','')
```

```python
df_ted.gender.value_counts()
```

```python
df_ted.gender.value_counts().plot('bar')
```

```python
df_ted.drop(df_ted[df_ted.gender == 'andy'].index, inplace=True)
```

```python
df_ted.gender.value_counts().plot('bar')
```

```python
df_ted.groupby('gender').mean().sort_values('views',ascending=False)['views'].plot('bar')
```

```python
round(df_ted.groupby('gender').mean().sort_values('duration', ascending=True)['duration']).plot('barh')
```

```python
#Hoe vaak per jaar praat male vs. female over onderwerp: health
df_health = pd.DataFrame(df_ted[(df_ted.gender == 'female')&(df_ted.tags.str.contains('health'))].groupby('year').count()['tags'])
df_health['male'] = df_ted[(df_ted.gender == 'male')&(df_ted.tags.str.contains('health'))].groupby('year').count()['tags']
df_health.columns = ['male','female']
df_health.plot(title='Aantal keer gesproken over health per jaar')
```
